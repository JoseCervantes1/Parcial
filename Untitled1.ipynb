```python
# Manejo de ciclos y validaciones

numeros = []
positivos = 0
negativos = 0
suma_positivos = 0

while True:
    try:
        numero = int(input("Ingrese un número entero (0 para terminar): "))
        if numero == 0:
            break

        numeros.append(numero)

        if numero > 0:
            positivos += 1
            suma_positivos += numero
        elif numero < 0:
            negativos += 1
    except ValueError:0
        print("Entrada inválida. Por favor, ingrese un número entero.")

print(f"\nSe ingresaron {positivos} números positivos.")
print(f"Se ingresaron {negativos} números negativos.")

if positivos > 0:
    promedio_positivos = suma_positivos / positivos
    print(f"El promedio de los números positivos es: {promedio_positivos:.2f}")
else:
    print("No se ingresaron números positivos.")

print("Todos los números ingresados (excepto el 0 final):", numeros)
```

```python
# Funciones + cadenas + booleanos

def analizar_texto(texto):
    longitud = len(texto)

    vocales = 0
    for letra in texto.lower():
        if letra in "aeiou":
            vocales += 1

    mayusculas = 0
    for letra in texto:
        if letra.isupper():
            mayusculas += 1
    texto_limpio = texto.replace(" ", "").lower()
    es_palindromo = texto_limpio == texto_limpio[::-1]

    return {
        "longitud": longitud,
        "vocales": vocales,
        "mayusculas": mayusculas,
        "es_palindromo": es_palindromo
    }

frase = input("Ingrese una frase: ")
resultado = analizar_texto(frase)
print(resultado)
```

```python
# Listas + ciclos for + cálculos

N = int(input("Ingrese un número: "))

numeros = list(range(1, N + 1))

cuadrados = [x**2 for x in numeros]
suma_cuadrados = sum(cuadrados)
maximo = max(cuadrados)
minimo = min(cuadrados)
multiplos_de_3 = 0
for num in cuadrados:
    if num % 3 == 0:
        multiplos_de_3 += 1
print("Lista de números:", numeros)
print("Lista de cuadrados:", cuadrados)
print("Suma de los cuadrados:", suma_cuadrados)
print("Valor máximo:", maximo)
print("Valor mínimo:", minimo)
print("Cantidad de cuadrados múltiplos de 3:", multiplos_de_3)
```

```python
# Diccionarios + funciones + ciclos while

def agregar_producto(inventario):
    nombre = input("Ingrese el nombre del producto: ").strip()
    if nombre in inventario:
        print("El producto ya existe.")
    else:
        cantidad = int(input("Ingrese la cantidad: "))
        inventario[nombre] = cantidad
        print("Producto agregado correctamente.")


def actualizar_cantidad(inventario):
    nombre = input("Ingrese el nombre del producto: ").strip()
    if nombre in inventario:
        cantidad = int(input("Ingrese la nueva cantidad: "))
        inventario[nombre] = cantidad
        print("Cantidad actualizada.")
    else:
        print("El producto no existe.")


def mostrar_inventario(inventario):
    if not inventario:
        print("El inventario está vacío.")
    else:
        print("  Inventario  ")
        for producto, cantidad in inventario.items():
            print(f"{producto}: {cantidad}")


def eliminar_producto(inventario):
    nombre = input("Ingrese el nombre del producto a eliminar: ").strip()
    if nombre in inventario:
        del inventario[nombre]
        print("Producto eliminado.")
    else:
        print("El producto no existe.")

def menu():
    inventario = {}

    while True:
        print("   MENÚ   ")
        print("1. Agregar producto")
        print("2. Actualizar cantidad")
        print("3. Mostrar inventario")
        print("4. Eliminar producto")
        print("5. Salir")

        opcion = input("Seleccione una opción: ")

        if opcion == "1":
            agregar_producto(inventario)
        elif opcion == "2":
            actualizar_cantidad(inventario)
        elif opcion == "3":
            mostrar_inventario(inventario)
        elif opcion == "4":
            eliminar_producto(inventario)
        elif opcion == "5":
            print("Saliendo del programa...")
            break
        else:
            print("Opción inválida. Intente de nuevo.")
menu()
```

```python
# Archivo final: funciones + listas + lógica compleja

def evaluar_notas(lista_notas):
    promedio = sum(lista_notas) / len(lista_notas)
    mayor = max(lista_notas)
    menor = min(lista_notas)

    aprobados = 0
    reprobados = 0

    for nota in lista_notas:
        if nota >= 3.0:
            aprobados += 1
        else:
            reprobados += 1

    return {
        "promedio": promedio,
        "mayor": mayor,
        "menor": menor,
        "aprobados": aprobados,
        "reprobados": reprobados
    }


cantidad = int(input("¿Cuántas notas va a ingresar?: "))

notas = []
for i in range(cantidad):
    nota = float(input(f"Ingrese la nota {i+1}: "))
    notas.append(nota)

resultado = evaluar_notas(notas)

print("   Resultados   ")
print(resultado)
```
