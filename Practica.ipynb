```python
cadena = input ("ingrese una cadena de texto:")
cadena = cadena.lower()
a = 0
e = 0
i = 0
o = 0
u = 0
for letra in cadena:
  if letra == "a":
    a += 1
  elif letra == "e":
    e += 1
  elif letra == "i":
    i += 1
  elif letra == "o":
    o += 1
  elif letra == "u":
    u += 1
print(f"Cantidad de 'a': {a}")
print(f"Cantidad de 'e': {e}")
print(f"Cantidad de 'i': {i}")
print(f"Cantidad de 'o': {o}")
print(f"Cantidad de 'u': {u}")
```

```python
cadena= input("Ingrese una cadena de texto: ")

cadena_modificada = []
for caracter in cadena:
    if caracter.isalpha():
        cadena_modificada.append(caracter.upper())
    else:
        cadena_modificada.append(caracter.lower())

resultado = "".join(cadena_modificada)
print(f"Cadena transformada: {resultado}")


print("\nCaracteres individuales de la cadena transformada:")
for i, char in enumerate(resultado):
    print(f"Caracter {i+1}: '{char}'")
```

```python
cadena = input("Ingrese una cadena de texto: ")


cadena_procesada = "".join(filter(str.isalnum, cadena)).lower()


cadena_invertida = cadena_procesada[::-1]


if cadena_procesada == cadena_invertida:
    print(f"La cadena '{cadena}' ES un palíndromo.")
else:
    print(f"La cadena '{cadena}' NO es un palíndromo.")
```
